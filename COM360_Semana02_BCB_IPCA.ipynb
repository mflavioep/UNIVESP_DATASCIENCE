* Foi escolhido o IPCA como objeto de estudo.
* O IPCA mede percentualmente a evolução do custo de vida das famílias brasileiras que ganham entre 1 e 40 salários mínimos e neste índice são considerados 9 grupos: alimentação, comunicação, residência, educação, despesas pessoais, saúde, habitação, transporte e vestuário.
Ele também é usado pelo Banco Central do Brasil (Bacen), para perseguir a meta de inflação. Este indicador é calculado mensalmente pelo Instituto Brasileiro de Geografia e Estatística (IBGE). A coleta de dados ocorre do primeiro ao último dia de cada mês nas capitais brasileiras.

Desta forma deve-se limpar a base de outros indicadores, deixando somente o IPCA.

```python
inflacao = inflacao[inflacao.Indicador == "IPCA"] 
```

```python
inflacao['Indicador'].value_counts()
```

```python
inflacao.head(10)
```

* Todos os registros são para o IPCA, portanto, esta coluna já não tem mais importância na base;
* As colunas Suavizada, numeroRespondentes, baseCalculo e Mediana também 

```python
inflacao['Media'].value_counts()
```

Limpando as colunas

```python
inflacao.drop(['Indicador', 'Mediana', 'numeroRespondentes', 'baseCalculo', 'Suavizada'], axis='columns', inplace=True)
```

```python
inflacao.head(10)
```

```python
inflacao.describe()
```

```python
import matplotlib.pyplot as plt
import datetime as dt
```

```python
inflacao['Data'] = pd.to_datetime(inflacao['Data'])
```

```python
inflacao.dtypes
```

```python
inflacao['Ano'] = inflacao['Data'].dt.year
```

```python
inflacao['Mes'] = inflacao['Data'].dt.month
```

```python
inflacao.head(2)
```

```python
inflacao.dtypes
```

```python
#Na primeira tentativa de conversão ocorreu erro de valores não informados.
inflacao.head(100)
```

```python
inflacao['Media'].fillna(method="ffill",inplace=True)
inflacao['DesvioPadrao'].fillna(method="ffill",inplace=True)
inflacao['Maximo'].fillna(method="ffill",inplace=True)
inflacao['Maximo'].fillna(method="ffill",inplace=True)
```

```python
#Ainda gerou erro de conversão
inflacao.head(500)
```

```python
#Os dados foram observados e possivelmente o erro de conversão é por causa da virgula em vez do ponto
#Existem opções para importar de forma correta, desde que o separador seja diferente de , (comma)
inflacao["Media"]=inflacao["Media"].str.replace(',','.')
inflacao["DesvioPadrao"]=inflacao["DesvioPadrao"].str.replace(',','.')
inflacao["Minimo"]=inflacao["Minimo"].str.replace(',','.')
inflacao["Maximo"]=inflacao["Maximo"].str.replace(',','.')
```

```python
inflacao.head(5)
```

```python
#Converte os tipos das colunas em questão
inflacao[["Media", "DesvioPadrao", "Minimo", "Maximo"]] = inflacao[["Media", "DesvioPadrao", "Minimo", "Maximo"]].apply(pd.to_numeric)
```

```python
inflacao.dtypes
```

```python
inflacao.plot(x='Data', y=['Media','Minimo','Maximo'])
```
